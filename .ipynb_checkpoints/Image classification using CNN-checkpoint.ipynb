## Image Classification with CNN


Dataset: Food-101, With 101 Classes with 1000 unique images but working with all this images will take time 
so for this project we have used only two classes

Classes:
1. Pizza
2. Chicken-Wings

And each classes contain 10000 unique images
we have seperated all the data into training and test data set with ration of 70-30 i.e 70% for training dataset and 30% for test set i.e alltogether 750 for each classes for training dataset and 250 for each testset.

## Section -1 initial set up and importing all the required library
    

```python
import matplotlib.pyplot as plt
from keras.models import Sequential  # Our sequential network or it can be 
from keras.layers import Conv2D # images are all 2d videos can be 3d
from keras.layers import MaxPooling2D #Pooling for downsampling
from keras.layers import Flatten #sinle linear  continious vector
from keras.layers import Dense # for neurral network
from keras.preprocessing.image import ImageDataGenerator #to work with images

```

#  Step 2 Conviolution/Pooling/Flattening/Dense

```python
classifier = Sequential() #Initialize the CNN

# CONV2D(number of filters,shape of the choosen image,input image type,activation funation)
classifier.add(Conv2D(32,(3,3),input_shape = (64,64,3),activation='relu')) #Only one CONV LAYER
classifier.add(MaxPooling2D(pool_size = (2,2))) # choosen max_pooling with 2 * 2
classifier.add(Flatten())

#hidden layer
classifier.add(Dense(units = 1,activation='sigmoid')) #it's a binary classification so choosed Sigmodi

classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])

```

### Optimizer = 'adam'

```python
classifier.summary()
```

# Step 3 Fitting the images with CNN
* Preprocessing

```python
train_datagen = ImageDataGenerator(rescale = 1./255, shear_range = 0.2, zoom_range = 0.2, horizontal_flip = True)
test_datagen = ImageDataGenerator(rescale=1./255)
```

```python
training_set = train_datagen.flow_from_directory(r'C:\Users\Asus\Desktop\AI\Dataset\training_set',target_size = (64,64),batch_size =32,class_mode = 'binary')
```

```python
test_set = train_datagen.flow_from_directory(r'C:\Users\Asus\Desktop\AI\Dataset\test_set',target_size = (64,64),batch_size =32,class_mode = 'binary')
```

### Images for test case

```python
from PIL import Image
sample_training_image,_ = next(training_set)
## Mapping images in the form of a grid 1 row and 5 column
def plotImages(images):
    fig,axes = plt.subplots(1,5,figsize=(32,32))
    for img,ax in zip(images,axes):
        ax.imshow(img)
        ax.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
plotImages(sample_training_image[:5])
```

# Model Training for classification

```python
classifier.fit(training_set,steps_per_epoch=len(training_set)//32,epochs=15,validation_data = test_set,validation_steps=len(test_set)//32)
```

# Predecting
## Predicting if the given image is pizza or chicken wings


```python
import numpy as np
from keras.preprocessing import image


test_image = image.load_img(r'C:\Users\Asus\Desktop\AI\Dataset\test_pizza.jpg',target_size = (64,64))
test_image = image.img_to_array(test_image)
test_image = np.expand_dims(test_image,axis=0)
result = classifier.predict(test_image)
training_set.class_indices

if result[0][0] == 1:
    prediction = 'pizza'
else:
    prediction = 'chicken wings'
print("Your image is: ", prediction)
```
